```python
schema = StructType([
    StructField("comment", StringType(), True),
    StructField("name", StringType(),True),
    StructField("star", IntegerType(),True),
    StructField("user_name", StringType(), True)
])
```

```python
df = spark.read.csv(r"./data/output.csv", header=True, schema=schema)
```

```python
df.show()
```
